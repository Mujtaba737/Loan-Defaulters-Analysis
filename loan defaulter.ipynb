2. Data Cleaning

```python
# calculating percentage of null values for each column

null_cols = app_df.isna().sum().to_frame().reset_index().rename(columns={'index':'col_name', 0:'null_count'})
null_cols['null_pct'] = (null_cols['null_count']/app_df.shape[0])*100
```

```python
null_cols.sort_values(by='null_pct', ascending=False).head()
```

```python
# dropping columns that have null value % > 40

drop_cols = null_cols.query('null_pct >= 40')['col_name'].to_list()
app_df = app_df.drop(columns=drop_cols)
app_df.shape
```

```python
app_df.head()
```

```python
# analyzing all flag columns

flag_cols = [col for col in app_df.columns if col.startswith('flag')]
flag_cols
```

```python
flag_target = app_df[flag_cols + ['target']]
flag_target.head()
```

```python
# replacing str values in flag_own_car and flag_own_realty columns with int

flag_target.loc[:,('flag_own_car', 'flag_own_realty')] = flag_target.loc[:,('flag_own_car', 'flag_own_realty')].replace({'Y':'1', 'N':'0'})
flag_target = flag_target.astype({'flag_own_car':'int64','flag_own_realty':'int64'})
```

```python
# countplot of each flag column

plt.figure(figsize=(25,25))

for index, col_name in enumerate(flag_cols):
    plt.subplot(7, 4, index+1)
    plt.subplots_adjust(wspace=0.35, hspace=0.65)
    ax = sns.countplot(data=flag_target, x=col_name, hue='target')
    ax.set(title=f'countplot of {col_name}')
```

```python
flag_corr_df = app_df[flag_cols[:8] + ['target']]
```

```python
# correlation of flag columns with target

corr_df = round(flag_corr_df.corr(numeric_only=True),2)
corr_df.head()
```

```python
# heatmap of correlation of  flag columns with target

plt.figure(figsize=(10,5))
ax = sns.heatmap(corr_df, cmap='coolwarm', linewidths=0.5, annot=True)
ax.set(title='Heatmap')
plt.show()
```

```python
# dropping flag columns as they have low correlation to target

app_df = app_df.drop(columns=flag_cols)
app_df.shape
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(data=round(app_df[['ext_source_2','ext_source_3','target']].corr(),2), cmap='coolwarm', linewidths=0.5, annot=True)
plt.show()
```

```python
# dropping ext_source columns as they have low correlation to target

app_df = app_df.drop(columns=['ext_source_2','ext_source_3'])
app_df.shape
```

```python
amt_cols = app_df.columns[app_df.columns.str.startswith('amt') & ~app_df.columns.str.endswith('range')]
amt_cols
```

```python
range_cols = app_df.columns[app_df.columns.str.endswith('range')]
range_cols
```

```python
# dataframe consisting of only numerical data

num_data = app_df[num_cols]
num_data.head()
```

```python
num_data.groupby(['target']).size()/num_data.shape[0]*100
```

```python
# dataframe consisting of only defaulters data

defaulters = num_data[num_data['target']==1].drop(columns='target')
defaulters.head()
```

```python
# dataframe consisting of only non-defaulters data

repayers = num_data[num_data['target']==0].drop(columns='target')
repayers.head()
```

```python
# average of all numerical columns by target

plt.figure(figsize=(20,40))
for index, col in enumerate(num_cols[2:]):
    data = app_df.groupby('target')[col].mean().reset_index()
    
    plt.subplot(len(num_cols[2:]),3,index+1)
    plt.subplots_adjust(wspace=0.4, hspace=0.4)
    ax = sns.barplot(data=data, x='target', y=col, hue='target',width=0.4)
    ax.set(title=f'Average of {col} by target')
```

```python
# kde plot of all amount columns

plt.figure(figsize=(20,12))

for idx, col in enumerate(amt_cols):
    plt.subplot(2,2,idx+1)
    plt.subplots_adjust(wspace=0.2,hspace=0.3)
    ax = sns.kdeplot(data=num_data, x=col, hue="target")
    ax.set(title=f'kde plot of {col}')
```

```python
# percent distribution of defaulters and non-defaulters by range columns

plt.figure(figsize=(15,25))
for index, col in enumerate(range_cols):
    rng_trgt = (app_df.groupby("target")[col].value_counts(normalize=True)*100).to_frame().reset_index().sort_values(by=['target',col]).reset_index(drop=True)
    rng_trgt = rng_trgt.rename(columns={'proportion':'percent'})
    
    plt.subplot(len(range_cols),1,index+1)
    plt.subplots_adjust(wspace=0.1, hspace=0.5)
    ax = sns.barplot(data=rng_trgt, x=col, y='percent', hue='target')
```

```python
# calculating default rate by name_contract_type

defaults_nct = app_df[app_df['target']==1].groupby(['name_contract_type']).size()
total_nct = app_df.groupby(['name_contract_type']).size()
default_rate_nct = (defaults_nct/total_nct*100).to_frame().reset_index().rename(columns={0:'default_percent'})
default_rate_nct
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data=app_df, x="name_contract_type", hue='target', stat='percent')
plt.title('Percentage of name_contract_type')
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.barplot(data=default_rate_nct,x="name_contract_type", y='default_percent',width=0.5)
plt.title('Default rate by name_contract_type')
plt.show()
```

```python
# percent distribution of loans and default rates for all object columns

plt.figure(figsize=(25,60))
for index, col in enumerate(obj_cols):
    defaults = app_df[app_df["target"]==1].groupby([col]).size()
    total = app_df.groupby([col]).size()
    default_rate = (defaults/total*100).to_frame().reset_index().rename(columns={0:'default_percent'})
    
    plt.subplot(len(obj_cols),2,index+index+1)
    plt.subplots_adjust(wspace=0.25, hspace=0.35)
    ax = sns.countplot(data=app_df.sort_values(by=col), x=col, stat='percent',
                       hue=col, width=0.3 if index in (0,1) else 0.8)
    ax.set(title=f'Percentage distribution of loans by {col}')
    if index >= len(obj_cols) - 2:
        plt.xticks(rotation=90)
    
    plt.subplot(len(obj_cols),2,index+index+2)
    ax = sns.barplot(data=default_rate.sort_values(by=col),x=col, y='default_percent',
                     hue=col, width=0.3 if index in (0,1) else 0.8)
    ax.set(title=f'Default rate by {col}')
    if index >= len(obj_cols) - 2:
        plt.xticks(rotation=90)
```
